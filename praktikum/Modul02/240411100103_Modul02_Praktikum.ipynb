````python
def modus(data):
    dataDict = {}
    for dt in data:
        if dataDict.get(dt) == None:
            dataDict[dt] = 1
        else:
            dataDict[dt] += 1

    max_count = 0
    mode_value = None
    for key, value in dataDict.items():
        if value > max_count:
            max_count = value
            mode_value = key

    return mode_value

data_list = [1, 2, 2, 3, 4, 4, 4, 5]
print(f"Data: {data_list}")
print(f"Modus: {modus(data_list)}")
````

* **(b) Matrix 2D**

 i. Buat fungsi create_matrix(rows,cols) untuk membuat matriks 2D dengan jumlah baris sebanyak rows dan jumlah kolom sebanyak cols, dengan return value berupa matriks yang sudah dibuat, seperti yang dicontohkan pada Gambar 2a atau Gambar 2b
  Screenshot 2025-03-21 204234.png

````python
def create_matrix(rows, cols):
    matriks = []
    for i in range(rows):
        temp = []
        for j in range(cols):
            strin = 'Data matriks [%d,%d] : ' % (i, j)
            data = int(input(strin))
            temp.append(data)
        matriks.append(temp)
    return matriks
````

````python
def display_matrix(ma):
    for i in range(len(ma)):
        for j in range(len(ma[i])):
            print(ma[i][j], end=' ')
        print()

hasil = create_matrix(2, 3)
display_matrix(hasil)
````

  * iii. Buat fungsi untuk menjumlahkan dua buah matriks add_matrix(a,b), yaitu matriks a dan matriks b, dengan syarat kedua buah matriks memiliki ukuran yang sama, seperti yang dicontohkan pada Gambar 2c atau Gambar 2d



````python
def add_matrix(a, b):
    if len(a) != len(b):
        print("error  --- size of both matrix is not equal\nerror matrix")
        return None
    for row in range(len(a)):
        if len(a[row]) != len(b[row]):
            print("error  --- size of both matrix is not equal\nerror matrix")
            return None

    result = []
    for i in range(len(a)):
        temp = []
        for j in range(len(a[0])):
            temp.append(a[i][j] + b[i][j])
        result.append(temp)
    return result

print("\nInput untuk matriks pertama:")
mat1 = create_matrix(2, 3)
print("\nInput untuk matriks kedua:")
mat2 = create_matrix(2, 2)
print("\n")
display_matrix(mat1)
print("\n")
display_matrix(mat2)
print("\n")
result = add_matrix(mat1, mat2)
if result is not None:
    display_matrix(result)
````


  * iv. Buat fungsi untuk mengalikan dua buah matriks mul_matrix(a,d, yaitu matriks a dan matriks b, dengan syarat perkalian matriks (jumlah kolom matriks pertama sama dengan jumlah baris matriks kedua), seperti yang dicontohkan pada Gambar 2e atau Gambar 2f
  Screenshot 2025-03-21 204434.png

````python
def mul_matrix(a, b):
    if len(a[0]) != len(b):
        print(f"error --- require {len(a[0])} but found {len(b)}\nerror matrix")
        return None

    result = []
    for i in range(len(a)):
        temp = []
        for j in range(len(b[0])):
            sum = 0
            for k in range(len(b)):
                sum += a[i][k] * b[k][j]
            temp.append(sum)
        result.append(temp)
    return result

print("\nInput untuk matriks pertama:")
mat1 = create_matrix(2, 3)
print("\nInput untuk matriks kedua:")
mat2 = create_matrix(3, 2)
print("\n")
display_matrix(mat1)
print("\n")
display_matrix(mat2)

print("\n")
result = mul_matrix(mat1, mat2)
if result is not None:
    display_matrix(result)
````
